```python
train_order_details = orders.join(train_order.set_index("order_id"),on = "order_id")
```

```python
prior_order_details = orders.join(prior_order.set_index("order_id"),on = "order_id")
```

```python
train_order_details = train_order_details[train_order_details["eval_set"] == "train"]
```

```python
prior_order_details = prior_order_details[prior_order_details["eval_set"] == "prior"]
```

```python
prior_order_details = prior_order_details.dropna()
```

```python
prior_order_details.head()
```

```python
def weekdays(x):
    if x in [1,2,3,4,5]:
        return "Weekday"
    else:
        return "Weekend"
```

```python
prior_order_details["order_dow"] = prior_order_details["order_dow"].apply(lambda x : weekdays(x))
```

```python
prior_order_details.head()
```

```python
def hourofday(x):
    if x in [0,1,2,3,4,5,23,22]:
        return "Night"
    elif x in [6,7,8,9,10,11]:
        return "Morning"
    elif x in [12,13,14,15,16,17]:
        return "Afternoon"
    else:
        return "Evening"
```

```python
prior_order_details["order_hour_of_day"] = prior_order_details["order_hour_of_day"].apply(lambda x : hourofday(x))
```

```python
prior_order_details.head()
```

```python
train_order_details["order_dow"] = train_order_details["order_dow"].apply(lambda x : weekdays(x))
train_order_details["order_hour_of_day"] = train_order_details["order_hour_of_day"].apply(lambda x : hourofday(x))
```
